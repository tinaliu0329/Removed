### Parse the Text Using spaCy

```python
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
## The following codes refer an answer on Stack, See: 
##https://stackoverflow.com/questions/51685224/percentage-count-verb-noun-using-spacy

from __future__ import unicode_literals
import spacy,en_core_web_sm
from collections import Counter
nlp = en_core_web_sm.load()
c = Counter(([token.pos_ for token in nlp(StoneA)]))
sbase = sum(c.values())
for el, cnt in c.items():
    print(el, '{0:2.2f}%'.format((100.0* cnt)/sbase))
```

We should also notice that in English, pronoun makes up a significant part in POS(5.65%, 5.31% and 7.01%). In modern Chinese, however, pronoun is not distinguishable from normal noun; in classical Chinese, some words are more likely to be used in the position of pronoun and maybes the equivalence of noun. (For the pronoun used in Classical Chinese, see Michael A. Fuller's *An Introduction to Literary Chinese*, page 262.) In Chan's research, he defines a noun as "a word used for naming some person or thing" (Chan 80). Therefore, pronoun is overlooked when he was counting. Overall, I will omit pronoun when considering the occurance frequency of pronoun.

#### Adverb

#### Stative Verb

"A stative verb is one that describes a state of being, in contrast to a dynamic verb, which describes an action. The difference can be categorized by saying that stative verbs describe situations that are static or unchanging throughout their entire duration, whereas dynamic verbs describe processes that entail change over time" (Wikipedia https://en.wikipedia.org/wiki/Stative_verb). 
In English, stative verb is made up of a broad type of verbs. For examples of stative verb in English, see: https://www.thoughtco.com/stative-verb-1692139.

Chan's paper defines words that qualify the following rules as stative verb (Chan 93)
1. A stative verb can be preceded by a word meaning "very"
2. It does not take any object.
3. It describes a state of being and thus in meaning resemble an adjective in the predicative use in an English sentence.
Thus, stative verb in Chinese combines words that are from both the adjetive and verb categories. It is hard for us to use the result of the POS analysis to give the occurrence frequency of stative verb.

# Conclusion

From the above analysis, we can conclude that the three parts - A+B, C+D and E+F - are not significantly different from one another, indicating the "likeness/closeness between the vocabularies of texts" (Chan 117). As we clearly know that the English translation is done by two translators, why does the computational test fail to recognize the difference and give “the wrong result”?

Unlike Chan's result, which reveals that the extinct 120-chapter version is by the same author, my result can be interpreted in multiple ways. In the following section, I provide three hypotheses for explaining why I obtain this result. 
By providing possible explanations, I do not intend to prioritize one over the other. Instead, the purpose of listing these hypotheses is to evoke more meaningful questions about the computational analysis of style and the question of authorship.

### Hypothesis 1: Mastery of Translation

In this hypothesis, I assume that Chan’s method is performed perfectively and effectively gives an accurate result on authorship. 

As mentioned earlier, David Hawkes and John Minford work collaboratively on the translation. Minford actively participated in the translation of the first eighty chapters, and he translated the last forty chapters not too long after Hawkes completed the first eighty chapters. Is it possible that the two translators’ style is so similar that it can even pass the computational analysis? 

Accepting this hypothesis immediately questions the validity of Chan’s conclusion. If it is possible for Minford to translate in Hawkes’ style, could someone successfully imitate Cao Xueqin’s style and write the last forty chapters and pass the computational test? Therefore, the question of authorship seems to be proved in vain. Why should we care about authorship if the “another author” can perfectively write in the style of “the author”?

### Hypothesis 2: Problems in Chan’s Research

In this hypothesis, I assume the “wrong result” occurs due to the problems in Chan’s method. By repeating the wrong method, we can arrive only at the wrong result. 

Chan’s research assumes that different writers have their distinct styles, and the differences can be expressed as the way their sentences are constructed and therefore, reflected in the occurrence frequency of certain POS. Theoretically, it is impossible to deceive the POS test because no one write while calculating the POS makeup of their sentences, making POS test the best parameter for tagging an author’s writing style. 

Chan’s result, however, is challenged by Chao Kang’s research, who analyzed a different POS, the empty word (xuci). Empty words in Classical Chinese do not carry any semantic meaning but functions as syntactic markers. Because of the lack of sematic meaning, in many situations, empty words can be omitted according to the author’s will. Therefore, Chao considers empty words the best indicator of an author’s style, and by analyzing empty words, Chao ends by concluding the presence of different authors. 

Chan’s and Chao’s conclusions are equally compelling, as they followed a very similar method, that is, to statistically analyzes the characteristic of the text’s POS, but chose different POS for analyzing. This gives rise to the question that whose POS is better and how to determine a better POS.


### Hypothesis 3: Problems in My Research

In this hypothesis, I assume Chan’s method gives correct results, but my process fails to repeat his. 
Chan’s research is executed in more detail than mine. For instance, Chan takes account of “noun-condensed” word so that these words will only be counted once. The word “problem child,” for instance, is counted as two nouns by spaCy although it should be only counted once. 

Moreover, Chan performed more statistical analysis to test the statistical validity. In my research, I only measure the frequency of total occurrence of certain POS. In contrast, Chan collected the relative frequency of words and performed association test to testify the likeliness.
