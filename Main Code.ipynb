# Text Preparation

### Before the start


The Penguin Classic version of David Hawkes and John Minford's translation is downloaded from the internet in a pdf. file. Then, I used ZAMZAR, an online file convertor, to convert the pdf. file into one single txt. file.

The full text is manually broken down into three parts: chapter 1-40, chapter 41-80 and chapter 81-120. The book title and translator's information is manually removed.

### Import the text

```python
## import pandas and print the working directory

import pandas as pd
%pwd
```

I import the three-part text and name each StoneA (includes chapter 1-40), StoneB (includes 41-80) and StoneC (includes 81-120) respectively. Here, only StoneA is printed for the sake of space.

```python
StoneA = open("The story of the stone(1-40).txt").read()
StoneB = open("The story of the stone(41-80).txt").read()
StoneC = open("The story of the stone(81-120).txt").read()
print(StoneA)
```

### Count words

```python
## How many lines?
print(len(StoneA.split("\n")))
```

### Remove punctuation

The 32 basic English punctuation will be removed using the foloowing module.

```python
## import the module for punctuation removal
from string import punctuation
```

```python
## The following punctuation will be removed from the text:
print(punctuation)
```

```python
## Remove the punctuation from the above list while rendering all the capital letter into lower-case
for char in punctuation:
    StoneA = StoneA.lower().replace(char, "")
```

```python
## print the text again
print(StoneA)
```

#### Additional Adjustment

```python
## Remove punctuation that are not part of the punctuation module

StoneA = StoneA.replace("‘", "") 
StoneA = StoneA.replace("’", "") 
StoneA = StoneA.replace("“", "") 
```
